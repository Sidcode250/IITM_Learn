```python
##################### BINARY SEARCH #########################

l = [1,2,3,4,5,6,7,8,9]
def binary_search(v,l):
    if l == []:
        return False
    m = len(l)//2

    if v == l[m]:
        return m
    elif v < l[m]:
        return binary_search(v,l[:m])
    else:
        return binary_search(v,l[m+1:])

target = binary_search(23,l)
print(target)
```

```python
###################### SELECTION SORT ###########################   

def selection_sort(l):
    n = len(l)
    if n < 1:
        return False
    
    for i in range(n):
        mpos = i
        for j in range(i+1,n):
            if l[j] < l[mpos]:
                mpos = j
        l[i],l[mpos] = l[mpos], l[i]
    return(l)

l = [4,2,7,3,0,1]
target = selection_sort(l)
print(target)
```

```python
###################### INSERTION SORT #######################

def insertion_sort(l):
    n = len(l)
    if n < 1:
        return False

    for i in range(n):
        j = i
        while(j>0 and l[j] < l[j-1]):
            l[j],l[j-1] = l[j-1],l[j]
            j = j-1
    return l

l = [4,2,7,3,0,1]
target = insertion_sort(l)
print(target)
```

```python
####################### MERGE SORT ##############################

def merge(A,B):
    (m,n) = (len(A),len(B))
    (C, i, j, k) = ([], 0, 0, 0)
    while k < m+n:
        if i == m:
            C.extend(B[j:])
            k = k + (n-j)
        elif j == n:
            C.extend(A[i:])
            k = k + (n-i)
        elif A[i] < B[j]:
            C.append(A[i])
            k,i = k+1,i+1
        else:
            C.append(B[j])
            k,j = k+1,j+1
    return C

def mergesort(A):
    n = len(A)

    if n <= 1:
        return A
    
    L = mergesort(A[:n//2])
    R = mergesort(A[n//2:])

    B = merge(L,R)

    return B

Lis = [4,2,7,9,1,7,3]
t = mergesort(Lis)
print(t)
```
