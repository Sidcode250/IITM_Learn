```python
##################### BINARY SEARCH #########################

l = [1,2,3,4,5,6,7,8,9]
def binary_search(v,l):
    if l == []:
        return False
    m = len(l)/2

    if v == l[m]:
        return m
    elif v > l[m]:
        return binary_search(v,l[:m])
    else:
        return binary_search(v,l[m+1:])

target = binary_search(3,l)
print(target)
```
